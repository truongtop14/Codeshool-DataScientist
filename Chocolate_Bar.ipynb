```
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```
def find_in_country_list(my_string):
  if my_string != 'Unknown':
    max_conf = 0
    my_country = 'Unknown'
    my_string = my_string.lower()
    
    for country in country_list['Name']:
      temp_conf = similar(my_string[:min(10,len(country)) + 1], country.lower()[:min(10,len(country)) + 1])
      if max_conf < temp_conf:
        max_conf = temp_conf
        my_country = country
      
    return my_country.title()
  else:
    return my_string
    
      
    
```

```
chocolate['broad_bean_origin'] = chocolate['broad_bean_origin'].apply(find_in_country_list)
```
